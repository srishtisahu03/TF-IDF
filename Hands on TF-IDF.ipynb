```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('twitter.csv')
```

```python
df
```

```python
tweets = df['raw_tweet'].values
```

### Preprocessing

```python
stemmer = PorterStemmer()
```

```python
processed_tweets = []

for tweet in tweets:
    
#     print(tweet, '\n')
    
    tweet = re.sub(r'[^a-zA-Z]',' ',tweet)
    
#     print(tweet, '\n')
    
    tweet = tweet.lower()
#     print(tweet, '\n')
    
    tweet = ' '.join([word for word in tweet.split(' ') if not word in stopwords.words('english')])
#     print(tweet, '\n')
    
    tweet = [stemmer.stem(word) for word in tweet.split()]
#     print(tweet, '\n')
    
    tweet = [word for word in tweet if len(word) != 0]
#     print(tweet, '\n')
    
    tweet = ' '.join(tweet)
#     print(tweet, '\n')
    
    processed_tweets.append(tweet)
```

```python
processed_tweets;
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

enc_tweets = tfidf.fit_transform(processed_tweets)
```

```python
df = pd.DataFrame(enc_tweets.toarray(), columns = tfidf.get_feature_names_out())
df
```

### Saving the models and encoding

```python
df.to_csv('tfidf_enc.csv', index = False)
```
